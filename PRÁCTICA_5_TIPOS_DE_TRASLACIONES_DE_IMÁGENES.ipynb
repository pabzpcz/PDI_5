EJERCICIO 1 - TRASLACIÓN BÁSICA DE IMAGEN

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer imagen en blanco y negro
imagen = cv2.imread('CEREBRO.jpg', 0)

# Tamaño
filas, columnas = imagen.shape

# Traslación
M = np.float32([
    [1, 0, 50],
    [0, 1, 30]
])

# Aplicar transformación
trasladada = cv2.warpAffine(imagen, M, (columnas, filas))

# Mostrar imágenes
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(imagen, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(trasladada, cmap='gray')
plt.title('Trasladada')
plt.axis('off')

plt.show()
```

EJERCICIO 2 - TRASLACIÓN CON VALORES DECIMALES

EJERCICIO 3 - TRASLACIÓN + ROTACIÓN

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer imagen

imagen = cv2.imread('CEREBRO.jpg')

# Convertir colores
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Obtener tamaño
filas, columnas = imagen.shape[:2]

# TRASLACIÓN

M_tras = np.float32([
    [1, 0, 50],
    [0, 1, 30]
])

trasladada = cv2.warpAffine(
    imagen,
    M_tras,
    (columnas, filas)
)

# ROTACIÓN

# Centro de la imagen
centro = (columnas // 2, filas // 2)

# Rotar 45 grados
M_rot = cv2.getRotationMatrix2D(
    centro,
    45,
    1
)

rotada = cv2.warpAffine(
    trasladada,
    M_rot,
    (columnas, filas)
)

# Mostrar resultados

plt.figure(figsize=(15,5))

# Original
plt.subplot(1,3,1)
plt.imshow(imagen)
plt.title('Original')
plt.axis('off')

# Trasladada
plt.subplot(1,3,2)
plt.imshow(trasladada)
plt.title('Trasladada')
plt.axis('off')

# Rotada
plt.subplot(1,3,3)
plt.imshow(rotada)
plt.title('Traslación + Rotación')
plt.axis('off')

plt.show()
```

EJERCICIO 4 - TRASLACIÓN + ROTACIÓN + ESCALADO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer imagen

imagen = cv2.imread('CEREBRO.jpg')

# Convertir colores
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Obtener tamaño
filas, columnas = imagen.shape[:2]

# TRASLACIÓN

M_tras = np.float32([
    [1, 0, 40],
    [0, 1, 20]
])

trasladada = cv2.warpAffine(
    imagen,
    M_tras,
    (columnas, filas)
)

# ROTACIÓN

centro = (columnas // 2, filas // 2)

M_rot = cv2.getRotationMatrix2D(
    centro,
    45,
    1
)

rotada = cv2.warpAffine(
    trasladada,
    M_rot,
    (columnas, filas)
)

# ESCALADO

# Imagen más grande
escala_grande = cv2.resize(
    rotada,
    None,
    fx=1.5,
    fy=1.5
)

# Imagen más pequeña
escala_pequena = cv2.resize(
    rotada,
    None,
    fx=0.5,
    fy=0.5
)

# Mostrar resultados

plt.figure(figsize=(15,10))

# Original
plt.subplot(2,2,1)
plt.imshow(imagen)
plt.title('Original')
plt.axis('off')

# Rotada
plt.subplot(2,2,2)
plt.imshow(rotada)
plt.title('Traslación + Rotación')
plt.axis('off')

# Grande
plt.subplot(2,2,3)
plt.imshow(escala_grande)
plt.title('Escala 150%')
plt.axis('off')

# Pequeña
plt.subplot(2,2,4)
plt.imshow(escala_pequena)
plt.title('Escala 50%')
plt.axis('off')

plt.show()
```

EJERCICIO 5 - TRANSFORMACIÓN AFÍN Y RADIAL

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer imagen

imagen = cv2.imread('CEREBRO.jpg')

# Convertir colores
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Obtener tamaño
filas, columnas = imagen.shape[:2]

# TRANSFORMACIÓN AFÍN

# Puntos originales
pts1 = np.float32([
    [50, 50],
    [200, 50],
    [50, 200]
])

# Nuevas posiciones
pts2 = np.float32([
    [10, 100],
    [200, 50],
    [100, 250]
])

# Matriz afín
M_afin = cv2.getAffineTransform(pts1, pts2)

# Aplicar transformación
imagen_afin = cv2.warpAffine(
    imagen,
    M_afin,
    (columnas, filas)
)

# DISTORSIÓN RADIAL

# Crear mapas vacíos
map_x = np.zeros((filas, columnas), np.float32)
map_y = np.zeros((filas, columnas), np.float32)

# Centro de imagen
cx = columnas / 2
cy = filas / 2

# Recorrer cada píxel
for y in range(filas):
    for x in range(columnas):

        dx = x - cx
        dy = y - cy

        # Distancia al centro
        r = np.sqrt(dx**2 + dy**2)

        # Factor de distorsión
        factor = 1 + 0.00001 * r**2

        # Nuevas posiciones
        new_x = dx * factor + cx
        new_y = dy * factor + cy

        map_x[y, x] = new_x
        map_y[y, x] = new_y

# Aplicar distorsión
imagen_radial = cv2.remap(
    imagen,
    map_x,
    map_y,
    cv2.INTER_LINEAR
)

# Mostrar resultados

plt.figure(figsize=(15,5))

# Original
plt.subplot(1,3,1)
plt.imshow(imagen)
plt.title('Original')
plt.axis('off')

# Afín
plt.subplot(1,3,2)
plt.imshow(imagen_afin)
plt.title('Transformación Afín')
plt.axis('off')

# Radial
plt.subplot(1,3,3)
plt.imshow(imagen_radial)
plt.title('Distorsión Radial')
plt.axis('off')

plt.show()
```
